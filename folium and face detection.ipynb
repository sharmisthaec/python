```python
import cv2
```

```python
img=cv2.imread("newimage.png",1)
face=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
detect=face.detectMultiScale(img)
print(detect)
```

```python
for a,b,c,d in detect:
    img=cv2.rectangle(img,(a,b),(a+c,b+d),(0,255,0),3)
```

```python
cv2.imshow("Shar",img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```
