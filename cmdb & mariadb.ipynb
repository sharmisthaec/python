```python
import pandas
import pysnow
```

```python
df=pandas.read_csv("all_incidents.csv")
```

```python
df.head()
```

```python
# Extracting all the resource ids from the description of incidents
import re
r1_list=[]
description=list(df["description"])
key_word="ResourceId"

for i in range(len(description)):
    if key_word in description[i]:
        
        r1=re.findall("ResourceId:\s\w+",description[i])
        r1_list.extend(r1) 
        
    else:
        r1_list.append("Not Available")
r1_list
```

```python
#Extracting the hostname from the resource ids
#Spliting the resource ids
config_list=[]
for item in r1_list:
    config_list.extend(item.split(": "))

#Filtering the hosts
config_list=filter(lambda x: x!="ResourceId", config_list)

#Converting hosts to the existing dataframe column
df["Hosts"]=pandas.DataFrame(config_list)
config_list
```
