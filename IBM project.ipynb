```python
#checking duplicate entry
app_name="Instagram"
for row in android_apps:
    if row[0]==app_name:
        print row
        print("\n")

```

```python
#storing all duplicate entries

unique_apps=[]
duplicate_apps=[]

for row in android_apps:
    if row[0] in unique_apps:
        duplicate_apps.append(row[0])
       
        
    else:
        unique_apps.append(row[0])
       
        

    
print "Number of duplicate apps: ",len(duplicate_apps)
print(unique_apps[0:10])
```

```python
reviews_max = {}

for app in android_apps:
    name = app[0]
    n_reviews = float(app[3])
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
        
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
       
```

```python
print'Expected length:', len(reviews_max)
print'Original data set length:', len(android)
print'Actual length:', len(android) - 1181
```

```python
android_clean = []
already_added = []

for app in android_apps:
    name = app[0]
    n_reviews = float(app[3])
    
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(app)
        already_added.append(name)
```
