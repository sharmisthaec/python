```python
import pandas
import pysnow
```

```python
df=pandas.read_csv("incident.csv")
```

```python
df.head(10)
```

```python
# Extracting all the resource ids from the description of incidents
import re
r1_list=[]
description=list(df["description"])
key_word="ResourceId"

for i in range(len(description)):
    if key_word in description[i]:
        
        r1=re.findall("ResourceId:\s\w+",description[i])
        r1_list.extend(r1) 
        
    else:
        r1_list.append("Not Available")
r1_list
```

```python
#Extracting the hostname from the resource ids
#Spliting the resource ids
config_list=[]
for item in r1_list:
    config_list.extend(item.split(": "))

#Filtering the hosts
config_list=filter(lambda x: x!="ResourceId", config_list)

#Converting hosts to the existing dataframe column
df["Hosts"]=pandas.DataFrame(config_list)
config_list
```

```python
class_name_list=[]
c = pysnow.Client(instance='dev90840', user='admin', password='Ntklie@25a54321')

for item in config_list:
    response = c.resource(api_path='/table/cmdb_ci').get(query={'name': item}, stream=True)

    try :
       
        class_name=" ".join(response["sys_class_name"][8:].split("_")).encode("UTF-8").upper()
        class_name_list.append(class_name)
        
    except Exception :
        
        class_name_list.append("No Record")

        
```

```python
df["Class"]=pandas.DataFrame(class_name_list)
```
